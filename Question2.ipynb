```python
import pandas as pd #to load json data to dataframe and iterate,manipulate through json data
```

```python
#loads data to dataframe -> SearchData and displays first 100 of them
SearchData = pd.read_json('searches.json', lines=True)
SearchData.head(100)
```

```python
#creating a dictionary of users who were shown the design B 
#containing user id (uid) as key & 
#number of times the user used the search feature (search_count)as value
b_user = {}
for index, row in SearchData.iterrows():
    b = row['uid']
    if(b%2 != 0):
        b_user[b] = row['search_count']
b_user
    
```

```python
#creating a dictionary of users who were shown the design A 
#containing user id (uid) as key & 
#number of times the user used the search feature (search_count)as value
a_user = {}
for index, row in SearchData.iterrows():
    a = row['uid']
    if(a%2 == 0):
        a_user[a] = row['search_count']
a_user
```

```python
#Creating a cumulative count of the total number of times search bar was used by
#users viewing design B (b_search_count) & 
#number of times search bar was used by users viewing design A (a_search_count)
b_search_count = sum(b_user.values())
a_search_count = sum(a_user.values())
print("A search count",a_search_count)
print("B search count",b_search_count)
```

```python
#Finding the frequency of use of search bar in design B vs 
#frequency of use of search bar in design A
a_size = len(a_user)
b_size = len(b_user)
print(a_size)
print(b_size)
a_search_frequency = a_search_count/a_size
b_search_frequency = b_search_count/b_size
print("A search frequency",a_search_frequency)
print("B search frequency",b_search_frequency)
```

Inference :-
1. Users using design A have used search bar more number of times over users using design B
(266 vs 231)
2. Frequency of use of search bar in design A is higher than Frequency of use of search bar in design B
(0.79 vs 0.66)

1. Did more users use the search feature in the new design (B)? 
ANS : NO
2. Did users search more often in the new design (B)?
ANS : NO
